```python
Trasen_tree=[]
def traverse(file_dir=Trasen_dir,Relative_path=''):
    fs = os.listdir(file_dir)
    for f1 in fs:
        Trasen_tree.append(Relative_path+f1)
        tmp_path = os.path.join(file_dir,f1)
        if not os.path.isdir(tmp_path):
            pass
        else:
            if Relative_path=='':
                traverse(tmp_path,f1+'\\')
            else:
                traverse(tmp_path,Relative_path+f1+'\\')
traverse()
def check_file_and_get_path(file_name):
    if not file_name in Trasen_tree:
        logging.error(file_name+'不存在')
        return None
    else:
        return path.join(Trasen_dir,file_name)
```

```python
def GetFile_crypt(filename):
    path=check_file_and_get_path(filename)
    if not path:
        print('check_file_and_get_path')
        logging.error(filename+'不存在，无法计算MD5')
    elif not os.path.isfile(filename):
        logging.warn(filename+'不是文件，无法计算MD5')
    else:
        myhash =get_crypt(check_file_cfgs['module'])
        logging.debug('加密检验：'+check_file_cfgs['module'])
        with open(filename,'rb') as f:
            while True:
                b = f.read(8096)
                if not b:
                    break
                myhash.update(b)
        return myhash.hexdigest()
def chech_file_crypt(file_name, outline):
    return GetFile_crypt(file_name)== outline
```

```python
def get_str(section='',options='',config=None):
    if section!='':
        try:
            size_s=int(section)
        except ValueError as e:
            section=section
        else:
            if config!=None and size_s>=0 and size_s<len(config):
                section=config.sections()[size_s]
            else:
                section=section
    else :
        raise ValueError('错误使用')
    if options !='':
        try:
            size_o=int(options)
        except ValueError as e:
            options=options
        else:
            if config.has_section(section) and size_o>=0 and size_o<len(config[section]):
                options=config.get_str
            else:
                options=options
    return [section,options]
```
